### Overview

## Import Libraries

```python
#Import Standard Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
```

## Data Preprocessing

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Get Data Ready
data_file = "Data/features_3_sec.csv"
music_data = pd.read_csv(data_file)
music_data;
```

```python
# Remove unnecessary Fields
music_data.drop(labels=["filename","length"],inplace=True, axis=1)
music_data;
```

```python
label_mappings = dict(enumerate(music_data['label'].unique()))
label_mappings
```

```python
#Converting Label Column to Numerical Field
for k, v in label_mappings.items():
    music_data.loc[(music_data['label'] == v),'label_code'] = k
    
music_data
```

```python
music_data[['label','label_code']]
```

```python
# Split data into train/test datasets
from sklearn.model_selection import train_test_split
```

## Create Models 

### Random Forest (87.45% Accuracy)

#### Models Ran
 - Default Parameters (n_estimators=100, criterion=gini)
 - Accuracy (n_estimators = 100, criterion = gini) = 0.8635301968635302 
 - Accuracy (n_estimators = 1000, criterion = gini) = 0.8745412078745413 - Best
 - Accuracy (n_estimators = 10,000, criterion = gini) = 0.8702035368702036
 - Accuracy (n_estimators = 1000, criterion = entropy) = 0.8732065398732065
 - Accuracy (n_estimators = 100, criterion = entropy) = 0.8625291958625292
 - Accuracy (n_estimators = 100, criterion = log loss) = 0.8668668668668669
 - Accuracy (n_estimators = 1000, criterion = log loss) = 0.8705372038705372

```python
# Create Random Forest Model
# Default Parameters (n_estimators=100, criterion='gini'): Accuracy = 0.8635301968635302 
# n_estimators = 1000: Accuracy = 0.8735402068735402
# n_estimators = 10,000: Accuracy = 0.8702035368702036
ml_randomforest = RandomForestClassifier(n_estimators=1000)

# Train Random Forest
# Create X (Features)
rfX = music_data.drop(["label", "label_code"], axis=1)

# Create Y (Labels)
rfY = music_data['label']

rfX_train, rfX_test, rfY_train, rfY_test = train_test_split(rfX, rfY, test_size=0.3)
ml_randomforest.fit(rfX_train, rfY_train)

#ml_randomforest.score(rfX_train, rfY_train)

ml_randomforest.score(rfX_test, rfY_test)
```

```python
rf_preds = ml_randomforest.predict(rfX_test)

print(classification_report(rfY_test, rf_preds))
```

- n_estimators=1000, max_depth=6, learning_rate=0.3: 0.8958958958958959


```python
# Create and train XGBoost model w/ default parameters
ml_xgboost = xgb.XGBClassifier(n_estimators=1000, max_depth=6, learning_rate=0.3)

# Train XGBoost - Looks like XGBoost doesn't work with categorical data only numerical
# Create X (Features)
xgbX = music_data.drop(["label", "label_code"], axis=1)

# Create Y (Labels)
xgbY = music_data['label_code']

xgbX_train, xgbX_test, xgbY_train, xgbY_test = train_test_split(xgbX, xgbY, test_size=0.3)
ml_xgboost.fit(xgbX_train, xgbY_train)

xgb_preds = ml_xgboost.predict(xgbX_test)

ml_xgboost.score(xgbX_test, xgbY_test)
```

```python
print(classification_report(xgbY_test, xgb_preds))
```
